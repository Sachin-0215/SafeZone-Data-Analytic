# Data Analysis of our Dataset

```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as pl
```

```python
df1 = pd.read_csv("newData.csv")
df1
```

```python
df1['Create Date/Time'] = pd.to_datetime(df1['Create Date/Time'], format="%Y-%m-%d %H:%M:%S")
```

# Split Date Day And Time in different columns

```python
df1['year']  = df1['Create Date/Time'].dt.year
df1['month'] = df1['Create Date/Time'].dt.month
df1['day']   = df1['Create Date/Time'].dt.day
df1['hour'] = df1['Create Date/Time'].dt.hour
df1['minute'] = df1['Create Date/Time'].dt.minute
df1['second'] = df1['Create Date/Time'].dt.second
df1['day_of_week']  = df1['Create Date/Time'].dt.dayofweek
df1.head(5)
```

```python
df1['DayOfWeek'] = df1['Create Date/Time'].dt.day_name()
df1
```

```python
df1.drop(['District', 'Caller Source'], axis = 1)   #dropped unncecessary columns from table
```

```python
df1['Create Date/Time'].max()-df1['Create Date/Time'].min()    #TOTAL DAYS
```

```python
df1['Dates'] = pd.to_datetime(df1['Create Date/Time']).dt.date
df1['Time'] = pd.to_datetime(df1['Create Date/Time']).dt.time
df1.drop(['Create Date/Time'],axis=1)
```

# Functions for filtering data Event Wise

```python
# filter by days of week and Event Type
def EventDaysInfo(EventType,EventDay):
      nEvent = df1[(df1["Event Type"] == EventType) & (df1['DayOfWeek']==EventDay)]
      print("Total Events of",EventType, "ocuured on ",EventDay, "is " ,len(nEvent))
```

```python
a1=len(df1[(df1['DayOfWeek']=='Monday')])
a2=len(df1[(df1['DayOfWeek']=='Tuesday')])
a3=len(df1[(df1['DayOfWeek']=='Wednesday')])
a4=len(df1[(df1['DayOfWeek']=='Thursday')])
a5=len(df1[(df1['DayOfWeek']=='Friday')])
a6=len(df1[(df1['DayOfWeek']=='Saturday')])
a7=len(df1[(df1['DayOfWeek']=='Sunday')])

plt.figure(figsize=(8,5))

ndays=[a1,a2,a3,a4,a5,a6,a7]
nocases = [697, 751, 737, 739, 733, 730, 736]
days = ["Monday","Tuesday","Wednesday","Thursday","Friday","Saturday","Sunday"]
plt.plot(days,nocases,"yo:",linewidth=3 ,markersize=10,label="Level of Cases")
plt.title("Day wise cases",fontsize=15) #changes fontsize of the text
plt.xlabel("Days",fontsize=13)        
plt.ylabel("Number of cases",fontsize=13)
plt.legend(loc=4) #set the location of the text
plt.grid(color='w', linestyle='-.', linewidth=5)
```
