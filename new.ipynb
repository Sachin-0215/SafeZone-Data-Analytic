```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df1 = pd.read_csv("newData.csv")
df1
```

```python
df1['Create Date/Time'] = pd.to_datetime(df1['Create Date/Time'], format="%Y-%m-%d %H:%M:%S")
```

# Split Date Day And Time in different columns

```python
df1['year']  = df1['Create Date/Time'].dt.year
df1['month'] = df1['Create Date/Time'].dt.month
df1['day']   = df1['Create Date/Time'].dt.day
df1['hour'] = df1['Create Date/Time'].dt.hour
df1['minute'] = df1['Create Date/Time'].dt.minute
df1['second'] = df1['Create Date/Time'].dt.second
df1['day_of_week']  = df1['Create Date/Time'].dt.dayofweek
df1.head(5)
```

```python
df1['DayOfWeek'] = df1['Create Date/Time'].dt.day_name()
df1
```

```python
df1.drop(['District', 'Caller Source'], axis = 1)   #dropped unncecessary columns from table
```

```python
df1['Create Date/Time'].max()-df1['Create Date/Time'].min()    #TOTAL DAYS
```

```python
df1['Dates'] = pd.to_datetime(df1['Create Date/Time']).dt.date
df1['Time'] = pd.to_datetime(df1['Create Date/Time']).dt.time
df1.drop(['Create Date/Time'],axis=1)
```

# No. of Events in Week Day by Event Type

```python
T_index = df1['Event Type'].value_counts()[:10].index

Event_Type = []
Monday = []
Tuesday = []
Wednesday = []
Thursday = []
Friday = []
Saturday = []
Sunday = []

for i in T_index:
    temp_df = (df1[df1['Event Type'] == i])['DayOfWeek'].value_counts(normalize=True)
    Event_Type.append(i)
    Monday.append(temp_df.Monday)
    Tuesday.append(temp_df.Tuesday)
    Wednesday.append(temp_df.Wednesday)
    Thursday.append(temp_df.Thursday)
    Friday.append(temp_df.Friday)
    Saturday.append(temp_df.Saturday)
    Sunday.append(temp_df.Sunday)

WeekEvent_df = pd.DataFrame({'Monday' : Monday,
                            'Tuesday': Tuesday,  
                            'Wednesday': Wednesday, 
                            'Thursday': Thursday, 
                            'Friday' : Friday, 
                            'Saturday' : Saturday, 
                            'Sunday' : Sunday
                            }, index=Event_Type)
```

```python
WeekEvent_df.plot(kind='bar', figsize=(20, 10), rot=20, stacked=True)
```

# No. of Events in Hour by Event Type

```python
T_index = df1['Event Type'].value_counts()[:10].index

Event_Type = []

hour_lst = {}
for i in range(24):
    hour_lst[i] = []

for i in T_index:
    temp_df = (df1[df1['Event Type'] == i])['hour'].value_counts(normalize=True)
    Event_Type.append(i)    
    for i in hour_lst.keys():
        try:
            hour_lst[i].append(temp_df[i])
        except:
            hour_lst[i].append(0)

HourEvent_df = pd.DataFrame(hour_lst,  index=Event_Type)
HourEvent_df.plot(kind='bar', figsize=(20, 10), rot=20, stacked=True)
```
