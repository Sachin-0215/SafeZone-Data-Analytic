# Machine Learning on Sequence Data

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("newData.csv")
```

```python
df.head()
```

```python
df['Create Date/Time'] = pd.to_datetime(df['Create Date/Time'], format="%Y-%m-%d %H:%M:%S")
```

```python
df.info()
```

```python
# remove unusable fields
dw_mapping={
    0: 'Monday', 
    1: 'Tuesday', 
    2: 'Wednesday', 
    3: 'Thursday', 
    4: 'Friday',
    5: 'Saturday', 
    6: 'Sunday'
} 
df = df[['Event Type', 'Create Date/Time', 'Latitude', 'Longitude']]
df['date'] = df['Create Date/Time'].dt.date
df['time'] = df['Create Date/Time'].dt.time
df['month'] = df['Create Date/Time'].dt.month
df['day']   = df['Create Date/Time'].dt.day
df['hour'] = df['Create Date/Time'].dt.hour
df['minute'] = df['Create Date/Time'].dt.minute
df['day_of_week_name'] = df['Create Date/Time'].dt.dayofweek.map(dw_mapping)

df.head(5)
```

### No. of Events in Week Day by Event Type

```python
T_index = df['Event Type'].value_counts()[:10].index

Event_Type = []
Monday = []
Tuesday = []
Wednesday = []
Thursday = []
Friday = []
Saturday = []
Sunday = []

for i in T_index:
    temp_df = (df[df['Event Type'] == i])['day_of_week_name'].value_counts(normalize=True)
    Event_Type.append(i)
    Monday.append(temp_df.Monday)
    Tuesday.append(temp_df.Tuesday)
    Wednesday.append(temp_df.Wednesday)
    Thursday.append(temp_df.Thursday)
    Friday.append(temp_df.Friday)
    Saturday.append(temp_df.Saturday)
    Sunday.append(temp_df.Sunday)

WeekEvent_df = pd.DataFrame({'Monday' : Monday,
                            'Tuesday': Tuesday,  
                            'Wednesday': Wednesday, 
                            'Thursday': Thursday, 
                            'Friday' : Friday, 
                            'Saturday' : Saturday, 
                            'Sunday' : Sunday
                            }, index=Event_Type)
```

```python
WeekEvent_df
```

```python
WeekEvent_df.plot(kind='bar', figsize=(20, 10), rot=20, stacked=True)
```

### No. of Events in Hour by Event Type

```python
T_index = df['Event Type'].value_counts()[:10].index

Event_Type = []

hour_lst = {}
for i in range(24):
    hour_lst[i] = []

for i in T_index:
    temp_df = (df[df['Event Type'] == i])['hour'].value_counts(normalize=True)
    Event_Type.append(i)    
    for i in hour_lst.keys():
        try:
            hour_lst[i].append(temp_df[i])
        except:
            hour_lst[i].append(0)

HourEvent_df = pd.DataFrame(hour_lst,  index=Event_Type)
```

```python
HourEvent_df
```

```python
HourEvent_df.plot(kind='bar', figsize=(20, 10), rot=20, stacked=True)
```

# Find Sensitive Areas Algorithm

### F = No. of crime at single point
### R = No. of recent crime at single point
### N = No. of crime in neabour points

```python
df.head()
```

```python
lat = 26.876
long = 81.055
```

```python
# F = No. of crime at single point
F = len(df.loc[(df['Latitude'] == lat) & (df['Longitude'] == long)])
F
```

```python
from datetime import date, timedelta
today = date.fromisoformat('2021-06-30')

one_day = today-timedelta(days=1)
three_day = today-timedelta(days=3)
one_week = today-timedelta(days=7)
one_month = today-timedelta(days=29)

one_day, three_day, one_week, one_month
```

```python
### R = No. of recent crime at single point
R = len(df.loc[(df['Latitude'] == lat) & (df['Longitude'] == long) & (df['date'] >= one_month)])
R
```

```python
# N = No. of crime in neabour points
N = len(df.loc[(df['Latitude'] > lat-0.0011) & (df['Latitude'] < lat+0.0011) 
               & (df['Longitude'] > long-0.0011) & (df['Longitude'] < long+0.0011) 
              ])
N
```

```python
point = (F*0.4) + (R*0.4) + (N*0.2)
point
```

```python
def CalculatePoint(lat, long, recent_days):
    recent_time =  date.fromisoformat('2021-06-30')-timedelta(days=recent_days)
    F = len(df.loc[(df['Latitude'] == lat) & (df['Longitude'] == long)])
    R = len(df.loc[(df['Latitude'] == lat) & (df['Longitude'] == long) & (df['date'] >= recent_time)])
    N = len(df.loc[(df['Latitude'] > lat-0.0011) & (df['Latitude'] < lat+0.0011) 
               & (df['Longitude'] > long-0.0011) & (df['Longitude'] < long+0.0011)])
    Fw = 0.2
    Rw = 0.5
    Nw = 0.2
    return (F*Fw) + (R*Rw)+ (N*Nw)
    

CalculatePoint(26.876, 81.055, 29)
```

```python
df.head()
```

```python
df2 = df.loc[:,'Latitude': 'Longitude']
lat_long = [] 
for i in df2.values:
    lat_long.append(str(i[0])+','+str(i[1]))
df2['lat_long'] = lat_long
```

```python
len(df2['lat_long'].unique())
```

```python
df2.head()
```

```python
max_points = {
    'points' : [],
    'lat' : [],
    'long' : []
}

for i in df2['lat_long'].unique():
    lst = [float(k) for k in i.split(',')]
    max_points['points'].append(CalculatePoint(lst[0], lst[1], 3))
    max_points['lat'].append(lst[0])    
    max_points['long'].append(lst[1])
    
points_df = pd.DataFrame(max_points)
points_df = points_df.sort_values(by='points', ascending=False)
points_df.head()
```
